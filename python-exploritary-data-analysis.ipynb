```python
#Value counts looks for and counts unique values  and arranges from most to least occuring

df['Year_Introduced'].value_counts()

#Gives an idea of how many roller coasters were introduced and which year
```

```python
#Identify first 10 years that rollercoasters were introduced and plot
#using a bar chart.

df['Year_Introduced'].value_counts().head(10).plot(kind='bar',title = 'Top Years Coasters Introduced')
```

```python
# Save as a Matplotlib axis

ax = df['Year_Introduced'].value_counts().head(10).plot(kind='bar',title = 'Top Years Coasters Introduced')
```

```python
# Use axis to add additional info

ax = df['Year_Introduced'].value_counts().head(10).plot(kind='bar',title = 'Top 10 Years Coasters Introduced')

ax.set_xlabel('Year Introduced')
ax.set_ylabel('Count')
```

```python
# Idea of the Distribution and different bin sizes

df['Speed_Mph'].plot(kind='hist',bins=20, title='Coaster Speed(mph)')
```

```python
# Save this plot

ax=df['Speed_Mph'].plot(kind='hist',bins=20, title='Coaster Speed(mph)')

ax.set_xlabel('Speed(mph)')
```

```python
#Density plot

ax=df['Speed_Mph'].plot(kind='kde', title='Coaster Speed(mph)')

ax.set_xlabel('Speed(mph)')
```

# Step 4: Feature relationships

```python
#Scatter plot (Matplotlib)

df.plot(kind='scatter', x='Speed_Mph',y='height_ft',title='Coaster Speed vs Height')
plt.show()
```

```python
#Seaborn (hold shift and tab to see signature info for the scatterplot)

sns.scatterplot(x='Speed_Mph',y='height_ft',data=df,hue='Year_Introduced')
plt.show()
```

```python
#Comparing more than 2 features with the colour of dots re[resented by the type of material used.

sns.pairplot(df,vars=['Year_Introduced','Speed_Mph','height_ft','Inversions_Clean','Gforce'],hue='Type_Main')
plt.show()
```

```python
# Correlation of above data, dropped nil values

df_corr = df[['Year_Introduced','Speed_Mph','height_ft','Inversions_Clean','Gforce']].dropna().corr()
```

```python
df_corr
```

```python
#Heatmap using df_corr

sns.heatmap(df_corr, annot = True)
plt.show()
```
